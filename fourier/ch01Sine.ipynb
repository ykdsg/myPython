```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- 1. 参数设置 ---
sampling_rate = 500  # 采样率 (Hz)，即每秒采集500个点
duration = 2         # 信号持续时间 (秒)
n_samples = int(sampling_rate * duration) # 总采样点数

# 第一个波形的参数
freq1 = 2            # 频率 (Hz)
amp1 = 3             # 振幅

# 第二个波形的参数
freq2 = 10           # 频率 (Hz)
amp2 = 1.5           # 振幅
```

```python
# --- 2. 生成时间轴 ---
# 创建一个从0到2秒的时间数组，包含 n_samples 个点
t = np.linspace(0, duration, n_samples, endpoint=False)
# --- 3. 生成波形 ---
# 生成第一个正弦波
# 公式: A * sin(2 * pi * f * t)
# freq1: 波形的频率（frequency），单位为赫兹(Hz)，决定了波形每秒钟振动的次数
wave1 = amp1 * np.sin(2 * np.pi * freq1 * t)

# 生成第二个正弦波
wave2 = amp2 * np.sin(2 * np.pi * freq2 * t)

# 将两个波形相加，得到复合波形
composite_wave = wave1 + wave2
```

```python
# --- 4. 绘图 ---
# 设置图形的整体大小
plt.figure(figsize=(12, 8))

# 绘制第一个子图：Component 1
plt.subplot(3, 1, 1) # (3行, 1列, 第1个图)
plt.plot(t, wave1)
plt.title(f"Component 1: Frequency = {freq1}Hz, Amplitude = {amp1}")
plt.ylabel("Amplitude")
plt.grid(True)

# 绘制第二个子图：Component 2
plt.subplot(3, 1, 2) # (3行, 1列, 第2个图)
plt.plot(t, wave2)
plt.title(f"Component 2: Frequency = {freq2}Hz, Amplitude = {amp2}")
plt.ylabel("Amplitude")
plt.grid(True)

# 绘制第三个子图：Composite Wave
plt.subplot(3, 1, 3) # (3行, 1列, 第3个图)
plt.plot(t, composite_wave)
plt.title("Composite Wave (Wave 1 + Wave 2)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
# 假设你只拿到了第三幅图（那个复杂的复合波形），你完全不知道它的来源。傅里叶变换要做的，就是接收这个复合波形作为输入，然后分析并输出：“嘿，我发现你这个复杂的信号，主要是由一个频率为2Hz、振幅为3的波形，和一个频率为10Hz、振幅为1.5的波形组成的！”


# 自动调整子图间距，防止标题重叠
plt.tight_layout()
```
