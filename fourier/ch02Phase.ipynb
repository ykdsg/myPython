```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --- 1. 参数设置 ---
sampling_rate = 500
duration = 1
n_samples = int(sampling_rate * duration)

freq = 5      # 频率 (5 Hz)
amplitude = 1 # 振幅

# --- 2. 生成时间轴 ---
t = np.linspace(0, duration, n_samples, endpoint=False)
```

```python
# --- 3. 生成两个相位不同的波形 ---

# 波形A: 相位为 0
phase_A = 0
wave_A = amplitude * np.sin(2 * np.pi * freq * t + phase_A)

# 波形B: 相位为 pi/2 (90度)
# np.pi 是 numpy 库中预定义的 π 值
phase_B = np.pi / 2
wave_B = amplitude * np.sin(2 * np.pi * freq * t + phase_B)

# --- 4. 绘图 ---
plt.figure(figsize=(10, 6))
plt.plot(t, wave_A, label=f"Wave A (Phase = 0)")
plt.plot(t, wave_B, label=f"Wave B (Phase = $\pi/2$)") # 使用LaTeX语法显示π

# 让我们看看波形B是不是一个余弦波
# 创建一个真正的余弦波来对比
cosine_wave = amplitude * np.cos(2 * np.pi * freq * t)
plt.plot(t, cosine_wave, 'r--', label="Cosine Wave (for comparison)") # 红色虚线

plt.title("Effect of Phase Shift")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend() # 显示图例
plt.show()
```
