```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = 3 + 4j
print(z)  # 输出: (3+4j)
print(z.real) # 输出: 3.0
print(z.imag) # 输出: 4.0
# 获取模 (Magnitude/Amplitude)
print(np.abs(z)) # 输出: 5.0 (因为 sqrt(3*3 + 4*4) = 5)
# 获取辐角 (Angle/Phase)，单位是弧度
print(np.angle(z)) # 输出: 0.927... (即 arctan(4/3) 的结果)
```

```python
# 1. 定义我们的初始向量，表示为复数
# 向量 (4, 0) 对应复数 4 + 0j
v = 4 + 0j
```

```python
# 2. 定义旋转角度
# 90度，换算成弧度是 π/2
angle_rad = np.pi / 2

# 3. 用欧拉公式创建“旋转器”
# e^(i*φ) 在 numpy 中就是 np.exp(1j * φ)
# 虚数i的几何意义是逆时针旋转90度。在复平面上，实数轴是横轴，虚数轴是纵轴。一个复数乘以i，就相当于将对应的向量逆时针旋转90度。
rotator = np.exp(1j * angle_rad)

# 打印一下这个“旋转器”看看它是什么
# cos(π/2) + i*sin(π/2) = 0 + 1i
print(f"Rotator (e^(i*π/2)): {rotator}") # 会输出一个非常接近 0+1j 的数
```

```python
# 4. 将向量与旋转器相乘，得到旋转后的新向量
v_rotated = v * rotator
print(f"Original vector: {v}")
print(f"Rotated vector: {v_rotated}") # 会输出一个非常接近 0+4j 的数
```

```python
# --- 5. 绘图验证 ---
plt.figure(figsize=(6, 6))
ax = plt.gca() # Get current axes

# 设置坐标轴范围
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
# 画出坐标轴
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_aspect('equal', adjustable='box')

# 画出原始向量 (从原点到v)
ax.arrow(0, 0, v.real, v.imag, head_width=0.2, head_length=0.3, fc='blue', ec='blue', label="Original Vector (4+0j)")
# 画出旋转后向量 (从原点到v_rotated)
ax.arrow(0, 0, v_rotated.real, v_rotated.imag, head_width=0.2, head_length=0.3, fc='red', ec='red', label=f"Rotated Vector ({v_rotated.real:.1f}+{v_rotated.imag:.1f}j)")

plt.title("Vector Rotation using Complex Multiplication")
plt.grid(True)
plt.legend()
plt.show()
```
